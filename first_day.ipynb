```python
 #which product contributes the highest percentage of total revenue?
product_revenue = d.groupby('Item Name')['Final Total'].sum().reset_index()
total_revenue = product_revenue['Final Total'].sum()

product_revenue['Revenue_Percentage'] = (
    product_revenue['Final Total'] / total_revenue
) * 100
top_product = product_revenue.sort_values(
    by='Revenue_Percentage',
    ascending=False
).head(1)

top_product
```
